```python
from scipy import optimize
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return (-12*x**4*np.sin(np.cos(x)) - 18*x**3 + 5*x**2 + 10*x - 30)

x = np.linspace(48, 53, 1001)
y = f(x)

root1 = optimize.root(f, x0=48)
root2 = optimize.root(f, x0=53)
minf = optimize.minimize(f, x0=48, bounds=((48, 53), ))

#plt.figure(figsize=(7, 3))
plt.plot(x, y, label='f', zorder=1)
plt.scatter([root1.x], [f(root1.x)], 30, label='1 корень', color='#000000')
plt.scatter([root2.x], [f(root2.x)], 30, label='2 корень', color='#000000')
plt.scatter([minf.x], [f(minf.x)], 30, label='Вершина', zorder=2, color='#FF0000')
plt.legend()
plt.xlim([48, 53])
plt.grid()
```
